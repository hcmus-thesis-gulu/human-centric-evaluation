## 2. Comparisons

```python
# metric = 'cosine'
metric = 'euclidean'
```

```python
from sklearn import metrics
dist = metrics.pairwise.distance_metrics()
print(dist[metric])
```

```python
cmp_idxs = np.arange(1, len(reduced_scaled_embeddings) - 1)
print(len(cmp_idxs))
```

```python
from scipy import spatial, stats
def distance(x, y):
    result_dist = np.array([dist[metric]([a], [b])[0, 0]
                            for a, b in zip(x, y)])
    print(result_dist.shape)
    return result_dist
```

```python
left_reduced_scaled = distance(reduced_scaled_embeddings[cmp_idxs], reduced_scaled_embeddings[:-2])
right_reduced_scaled = distance(reduced_scaled_embeddings[cmp_idxs], reduced_scaled_embeddings[2:])
reduced_scaled_dist = (left_reduced_scaled + right_reduced_scaled) / 2
print(reduced_scaled_dist)
print(reduced_scaled_dist.shape)
```

```python
norm_reduced_scaled_dist = preprocessing.normalize(reduced_scaled_dist.reshape(-1, 1),
                                                   norm="max", axis=0)[:, 0]
print(norm_reduced_scaled_dist)
```

```python
sns.lineplot(x = cmp_idxs,
             y = norm_reduced_scaled_dist)
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
high_range = int(len(norm_reduced_scaled_dist) * 0.15)
high_idxs = np.argpartition(norm_reduced_scaled_dist,
                            -high_range)[-high_range:]
print(high_idxs)
```

```python
norm_reduced_scaled_cutoff = norm_reduced_scaled_dist[high_idxs].min()
print(norm_reduced_scaled_cutoff)
```

```python
sns.lineplot(x = cmp_idxs,
             y = norm_reduced_scaled_dist)
plt.axhline(y=norm_reduced_scaled_cutoff,
            linestyle='--', color='red')
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
import pandas as pd
df = pd.DataFrame({
    'x': cmp_idxs,
    'y': norm_reduced_scaled_dist
})
print(df.info())
```

```python
mean_smoothed_dist = df['y'].rolling(window=5,
                                     center=True).mean().values
std_smoothed_dist = df['y'].rolling(window=5,
                                    center=True).std().values
print(mean_smoothed_dist)
print(std_smoothed_dist)
```

```python
sns.lineplot(x = cmp_idxs,
             y = mean_smoothed_dist)
sns.lineplot(x = cmp_idxs,
             y = std_smoothed_dist)
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
orig_idxs = samples[cmp_idxs]
print(orig_idxs)
```

```python
def f_measure(x, y):
    inter = np.nansum(np.min([x, y], axis=0))
    left = np.nansum(x)
    right = np.nansum(y)

    precision = inter / left
    recall = inter / right

    return 2 * precision * recall / (precision + recall)
```

```python
def gt_prep(gt_summary):
    norm_gt_summary = preprocessing.normalize(gt_summary.reshape(-1, 1),
                                              norm="max", axis=0)[:, 0]

    return norm_gt_summary[orig_idxs]
```

```python
general_summary = (user_score > 0).sum(axis=1)
print(general_summary.shape)
```

```python
general_score = gt_prep(general_summary)
print(general_score.shape)
```

```python
sns.lineplot(x = cmp_idxs,
             y = general_score)
plt.axhline(y=norm_reduced_scaled_cutoff,
            linestyle='--', color='red')
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
sns.lineplot(x = cmp_idxs,
             y = mean_smoothed_dist)
sns.lineplot(x = cmp_idxs,
             y = general_score)
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
sns.lineplot(x = cmp_idxs[:-1],
             y = mean_smoothed_dist[1:] - mean_smoothed_dist[:-1])
sns.lineplot(x = cmp_idxs,
             y = general_score)
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
sns.lineplot(x = cmp_idxs,
             y = std_smoothed_dist)
sns.lineplot(x = cmp_idxs,
             y = general_score)
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
sns.lineplot(x = cmp_idxs[1:],
             y = std_smoothed_dist[1:] - std_smoothed_dist[:-1])
sns.lineplot(x = cmp_idxs,
             y = general_score)
sns.set(rc={'figure.figsize':(20, 10)})
```

```python
print(f_measure(norm_reduced_scaled_dist, general_score))
print(f_measure(mean_smoothed_dist, general_score))
print(f_measure(std_smoothed_dist, general_score))
```

```python
mean_f_scores = []
std_f_scores = []

for user_idx in range(user_score.shape[1]):
    mean_f_scores.append(f_measure(mean_smoothed_dist,
                                   gt_prep(user_score[:, user_idx])))
    std_f_scores.append(f_measure(std_smoothed_dist,
                                  gt_prep(user_score[:, user_idx])))

print(mean_f_scores)
print(std_f_scores)
```

```python
mean_gt_summary = user_score[:, mean_user]
mean_gt_score = gt_prep(mean_gt_summary)
print(np.unique(mean_gt_summary))
print(np.unique(mean_gt_score))
```

```python
# model = cluster.DBSCAN(eps=0.001, metric=metric, n_jobs=-1)
# model = cluster.OPTICS(max_eps=1, metric=metric, n_jobs=-1)
# model = cluster.AgglomerativeClustering(20)
birch_model = cluster.Birch(threshold=0.5, n_clusters=None)
# model = cluster.KMeans(10, n_init='auto')
# model = cluster.AffinityPropagation(random_state=42)
# get the subcluster labels from the BIRCH model
subcluster_labels = birch_model.fit_predict(reduced_scaled_embeddings)
```

```python
# create the connectivity matrix for the agglomerative clustering model
row = []
col = []
subcluster_dict = {}
embedding_shape = reduced_scaled_embeddings.shape

for i in range(embedding_shape[0]):
    if subcluster_labels[i] not in subcluster_dict:
        subcluster_dict[subcluster_labels[i]] = []
    subcluster_dict[subcluster_labels[i]].append(i)
for subcluster in subcluster_dict.values():
    for i in range(len(subcluster)):
        for j in range(i+1, len(subcluster)):
            row.append(subcluster[i])
            col.append(subcluster[j])
            row.append(subcluster[j])
            col.append(subcluster[i])
connectivity = sparse.csr_matrix((np.ones(len(row)), (row, col)),
                                 shape=(embedding_shape[0], embedding_shape[0])
                                 )
```

```python
# create the agglomerative clustering model
agg_model = cluster.AgglomerativeClustering(n_clusters=27, linkage='single', connectivity=connectivity)
```

```python
reduced_labels = agg_model.fit_predict(reduced_scaled_embeddings)
```
